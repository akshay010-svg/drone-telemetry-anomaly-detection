```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold
from sklearn.metrics import classification_report, confusion_matrix, f1_score, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import class_weight
import sys
import io

FILE_PATH = '/content/Fusion_Data.csv'

old_stderr = sys.stderr
sys.stderr = io.StringIO()
try:
    df = pd.read_csv(
        FILE_PATH,
        sep=',' 
    )

except FileNotFoundError:
    sys.stderr = old_stderr
    print(f"--- ERROR ---")
    print(f"File '{FILE_PATH}' not found. Please upload your file.")
    raise
except Exception as e:
    sys.stderr = old_stderr
    print(f"An error occurred loading the file: {e}")
    raise
finally:
    sys.stderr = old_stderr

print(f"Dataset loaded successfully: {df.shape[0]} rows")
TARGET = 'labels'

print("\n--- Cleaning Data & Selecting Features ---")

ALL_LOADED_FEATURES = [col for col in df.columns if col not in ['timestamp', 'labels']]

cols_to_drop = [
    'ErrYaw', 'MagZ', 
    'DesRoll', 'DesPitch', 'DesYaw' 
]

FEATURES = [col for col in ALL_LOADED_FEATURES if col in df.columns and col not in cols_to_drop]

df = df.ffill().bfill()

selector = VarianceThreshold(threshold=0.0)
selector.fit(df[FEATURES])

features_to_keep = df[FEATURES].columns[selector.get_support()]
features_to_drop = df[FEATURES].columns[~selector.get_support()]

if len(features_to_drop) > 0:
    print(f"Dropping {len(features_to_drop)} constant columns: {list(features_to_drop)}")

FEATURES = list(features_to_keep)
NUM_FEATURES = len(FEATURES)
print(f"Using {NUM_FEATURES} features for training: {FEATURES}")

X = df[FEATURES]
y = df[TARGET]

X_train_df, X_test_df, y_train_df, y_test_df = train_test_split(
    X, y,
    test_size=0.2,
    shuffle=True,
    stratify=y, 
    random_state=42
)
print(f"Training samples: {len(X_train_df)}, Testing samples: {len(X_test_df)}")

# 7. Scale Data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_df)
X_test = scaler.transform(X_test_df)
print("Data scaled successfully.")

# 8. Calculate Class Weights
class_weights = class_weight.compute_class_weight(
    'balanced',
    classes=np.unique(y_train_df),
    y=y_train_df
)
class_weight_dict = dict(enumerate(class_weights))
print(f"Calculated Class Weights: {class_weight_dict}")



# MODEL 1 - "TEACHER" (DNN)


def build_dnn_teacher_model(num_features, num_classes=5):
    inp = layers.Input(shape=(num_features,))
    x = layers.Dense(64, activation='relu')(inp)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x) # Higher dropout

    x = layers.Dense(32, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x) # Higher dropout

    out = layers.Dense(num_classes, activation='softmax')(x)

    model = keras.Model(inp, out, name="DNN_Teacher_Model")
    model.compile(
        loss='sparse_categorical_crossentropy',
        optimizer=keras.optimizers.Adam(learning_rate=0.001), # Standard LR
        metrics=['accuracy']
    )
    return model

print("\n--- Training the DNN 'Teacher' Model ---")
dnn_model = build_dnn_teacher_model(NUM_FEATURES, num_classes=5)
dnn_model.summary()

# Use EarlyStopping to find the best model and prevent overfitting
early_stopping = keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10, # Stop if val_loss doesn't improve for 10 epochs
    restore_best_weights=True # Keep the best version of the model
)

history_dnn = dnn_model.fit(
    X_train, y_train_df.values,
    epochs=100, # Train for longer, EarlyStopping will find the best
    batch_size=32,
    validation_data=(X_test, y_test_df.values),
    class_weight=class_weight_dict,
    callbacks=[early_stopping],
    verbose=1
)

print("\n--- Evaluating the DNN 'Teacher' Model ---")
y_pred_dnn_prob = dnn_model.predict(X_test)
y_pred_dnn = np.argmax(y_pred_dnn_prob, axis=1)
report_dnn = classification_report(y_test_df.values, y_pred_dnn, output_dict=True, zero_division=0)
print(classification_report(y_test_df.values, y_pred_dnn, zero_division=0))


# MODEL 2 - "HYBRID" (DNN + PHYSICS)


print("\n--- Evaluating Hybrid Model (DNN + Physics) ---")

def check_physics_rules_hybrid(data_df_raw, scaler_obj):
    try:
        # Check if ErrRP exists (it should, based on our new logic)
        if 'ErrRP' in data_df_raw.columns:
            # Find the original mean and std of 'ErrRP'
            errp_mean = scaler_obj.mean_[scaler_obj.feature_names_in_.tolist().index('ErrRP')]
            errp_std = scaler_obj.scale_[scaler_obj.feature_names_in_.tolist().index('ErrRP')]
            # Set a threshold at 2 standard deviations
            threshold = errp_mean + (2 * errp_std)

            # Apply rule to the *original unscaled* data
            rule = (data_df_raw['ErrRP'].abs() > threshold)
            physics_pred = np.where(rule, 3, 0) # Predict "Class 3" (arbitrary)
            return physics_pred.astype(int)
        else:
            print("Could not apply physics rule ('ErrRP' was not in features).")
            return np.zeros(len(data_df_raw), dtype=int)
    except Exception as e:
        print(f"Could not apply physics rule: {e}")
        return np.zeros(len(data_df_raw), dtype=int)

y_pred_model1 = y_pred_dnn
y_pred_physics = check_physics_rules_hybrid(X_test_df, scaler)

y_pred_hybrid = np.copy(y_pred_model1)
model_missed_physics_violation = (y_pred_model1 == 0) & (y_pred_physics != 0)
y_pred_hybrid[model_missed_physics_violation] = y_pred_physics[model_missed_physics_violation]

report_hybrid = classification_report(y_test_df.values, y_pred_hybrid, output_dict=True, zero_division=0)
print(classification_report(y_test_df.values, y_pred_hybrid, zero_division=0))



# MODEL 3 - "LIGHTWEIGHT" (KNOWLEDGE DISTILLATION)

def build_student_model(num_features, num_classes=5):
    inp = layers.Input(shape=(num_features,))
    x = layers.Dense(16, activation='relu')(inp) # Tiny
    x = layers.Dropout(0.3)(x)
    out = layers.Dense(num_classes, activation='softmax')(x)

    model = keras.Model(inp, out, name="Student_DNN")
    model.compile(
        loss='sparse_categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )
    return model

student_model = build_student_model(NUM_FEATURES, num_classes=5)
print("\n--- Student Model Architecture ---")
student_model.summary()

class Distiller(keras.Model):
    def __init__(self, student, teacher):
        super().__init__()
        self.teacher = teacher
        self.student = student

    def compile(self, optimizer, metrics, student_loss_fn, distillation_loss_fn, alpha=0.1, temperature=3):
        super().compile(optimizer=optimizer, metrics=metrics)
        self.student_loss_fn = student_loss_fn
        self.distillation_loss_fn = distillation_loss_fn
        self.alpha = alpha
        self.temperature = temperature

    def train_step(self, data):
        if len(data) == 3:
            X, y, sample_weight = data
        else:
            X, y = data
            sample_weight = None

        teacher_predictions = self.teacher(X, training=False)

        with tf.GradientTape() as tape:
            student_predictions = self.student(X, training=True)
            student_loss = self.student_loss_fn(y, student_predictions, sample_weight=sample_weight)
            distillation_loss = self.distillation_loss_fn(
                tf.nn.softmax(teacher_predictions / self.temperature),
                tf.nn.softmax(student_predictions / self.temperature),
            )
            loss = self.alpha * student_loss + (1 - self.alpha) * distillation_loss

        trainable_vars = self.student.trainable_variables
        gradients = tape.gradient(loss, trainable_vars)
        self.optimizer.apply_gradients(zip(gradients, trainable_vars))
        self.compiled_metrics.update_state(y, student_predictions, sample_weight=sample_weight)
        return {m.name: m.result() for m in self.metrics}

print("\n--- Training Student Model via Distillation ---")
distiller = Distiller(student=student_model, teacher=dnn_model)
distiller.compile(
    optimizer=keras.optimizers.Adam(),
    metrics=['accuracy'],
    student_loss_fn=keras.losses.SparseCategoricalCrossentropy(),
    distillation_loss_fn=keras.losses.KLDivergence(),
    alpha=0.1,
    temperature=10
)

distiller.fit(
    X_train, y_train_df.values,
    epochs=100,
    batch_size=32,
    class_weight=class_weight_dict,
    callbacks=[keras.callbacks.EarlyStopping(monitor='accuracy', patience=10, restore_best_weights=True)],
    verbose=1
)

print("\n--- Evaluating Student Model ---")
y_pred_student_prob = student_model.predict(X_test)
y_pred_student = np.argmax(y_pred_student_prob, axis=1)
report_student = classification_report(y_test_df.values, y_pred_student, output_dict=True, zero_division=0)
print(classification_report(y_test_df.values, y_pred_student, zero_division=0))



# FINAL COMPARISON
print("\n\n" + "="*30)
print("     FINAL MODEL COMPARISON")
print("="*30)

params_teacher = dnn_model.count_params()
params_hybrid = params_teacher
params_student = student_model.count_params()

f1_teacher = report_dnn.get('weighted avg', {}).get('f1-score', 0.0)
f1_hybrid = report_hybrid.get('weighted avg', {}).get('f1-score', 0.0)
f1_student = report_student.get('weighted avg', {}).get('f1-score', 0.0)

acc_teacher = report_dnn.get('accuracy', 0.0)
acc_hybrid = report_hybrid.get('accuracy', 0.0)
acc_student = report_student.get('accuracy', 0.0)

comparison_data = {
    "Model": ["1. Core (DNN Teacher)", "2. Hybrid (DNN + Physics)", "3. Lightweight (DNN Student)"],
    "Weighted F1-Score": [f1_teacher, f1_hybrid, f1_student],
    "Overall Accuracy": [acc_teacher, acc_hybrid, acc_student],
    "Model Size (Params)": [params_teacher, params_hybrid, params_student]
}
comparison_df = pd.DataFrame(comparison_data)

print(comparison_df.to_markdown(index=False, floatfmt=".3f"))

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(22, 6))

sns.heatmap(confusion_matrix(y_test_df.values, y_pred_dnn), annot=True, fmt='d', ax=ax1, cmap='Blues')
ax1.set_title(f'Model 1: Teacher (DNN)\nF1: {f1_teacher:.3f} | Size: {params_teacher}', fontsize=14)
ax1.set_xlabel('Predicted'); ax1.set_ylabel('Actual')

sns.heatmap(confusion_matrix(y_test_df.values, y_pred_hybrid), annot=True, fmt='d', ax=ax2, cmap='Greens')
ax2.set_title(f'Model 2: Hybrid (DNN + Physics)\nF1: {f1_hybrid:.3f} | Size: {params_hybrid}', fontsize=14)
ax2.set_xlabel('Predicted'); ax2.set_ylabel('Actual')

sns.heatmap(confusion_matrix(y_test_df.values, y_pred_student), annot=True, fmt='d', ax=ax3, cmap='Oranges')
ax3.set_title(f'Model 3: Student (Distilled)\nF1: {f1_student:.3f} | Size: {params_student}', fontsize=14)
ax3.set_xlabel('Predicted'); ax3.set_ylabel('Actual')

plt.suptitle('Final Model Comparison', fontsize=20)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
